```python
import pandas as pd
from sklearn.datasets import load_iris
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt
```

```python
iris = load_iris()
iris_df = pd.DataFrame(iris.data, columns = iris.feature_names)
iris_df['Species'] = iris.target
iris_df['Species'] = iris_df['Species'].map({0:"setosa", 1:"versicolor", 2:"verginica"})
```

```python
plt.figure(figsize = (15,5))
plt.scatter(iris_df['sepal length (cm)'], iris_df['sepal width (cm)'], c=iris.target)
```

## Let's see what KMeans think how they should be clustered

```python
iris_kmeans = KMeans(n_clusters=3).fit(iris_df[['sepal length (cm)', 'sepal width (cm)']])
```

```python
plt.figure(figsize = (15,5))
plt.scatter(iris_df['sepal length (cm)'], iris_df['sepal width (cm)'], c=iris_kmeans.labels_)
centers = pd.DataFrame(iris_kmeans.cluster_centers_, columns = ["X","Y"])
plt.scatter(centers.X, centers.Y,c="blue",marker="X")
```
