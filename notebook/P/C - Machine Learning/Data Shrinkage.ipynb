## Regularization
- Regularization is a way to scale features of different magnitude to similar magnitute so that a optimizer can converge easily and without extreme intercept while maintaining original relationship amongst each other.
- There are couple of regularization methods
    - **L1 Regularization:** Where each random variables are divided by the sum of the absolute values
    - **L2 Regularization:** Where each random variables are divided by the square root of the sum of the squares
    - **Min Max Regularization:** Where each random variables are shrunk to a scale of 0 and 1
    - **StandardScaler:** Scale around mean and standardize it. RV becomes standard normal variables
    

```python
import numpy as np
import pandas as pd
from sklearn.preprocessing import StandardScaler
from sklearn.preprocessing import Normalizer
import matplotlib.pyplot as plt
from IPython.display import display_html
from IPython.display import Markdown
import warnings
warnings.filterwarnings('ignore')
```

## DATA

```python
temp = pd.DataFrame([ [4,1,2,2], [1,3,5,3] , [2,2,5,3]])
temp.columns = ["A","B","C","D"]
temp
```

## L1 Regularization
Where each random variables are divided by the sum of the absolute values

Each random variables can be negative or positive
$$\mathbb{Y} = \{ y_n \in \mathbb{R} \}$$

$$\omega = \sum_{n=0}^{N} {|y_n|} $$

After regularization the random variables will be positive
$$\mathbb{Y} / \omega = \{ y_n \in \mathbb{R} : y_n > 0 \}$$

```python
print("*" * 83)
print(" Scikit learn does it WRONG where it applies it row wise which is definitely wrong")
print("*" * 83)
print("\n")
l1 = Normalizer(norm="l1")
print(l1.fit_transform(temp))
print("\n")

print("*" * 20)
print("   THE WRIGHT WAY")
print("*" * 20)
display_html(temp.div(temp.abs().sum()))
print("\n")

print("*" * 50)
print("         If you insist on scikit learn then")
print("*" * 50)
print("\n")
l1 = Normalizer(norm="l1")
print(l1.fit_transform(temp.T).T)
```

## L2 Regularization
Where each random variables are divided by the square root of the sum of the squares of each

Each random variables can be negative or positive
$$\mathbb{Y} = \{ y_n \in \mathbb{R} \}$$

$$\omega = \sqrt{\sum_{n=0}^{N} {y_n^2} }$$

After regularization the random variables will be positive
$$\mathbb{Y} / \omega = \{ y_n \in \mathbb{R} : y_n > 0 \}$$

```python
print("*" * 83)
print(" Scikit learn does it WRONG where it applies it row wise which is definitely wrong")
print("*" * 83)
print("\n")
l2 = Normalizer(norm="l2")
print(l2.fit_transform(temp))
print("\n")

print("*" * 83)
print("   THE WRIGHT WAY. Also luckily pandas does column wise computation by default")
print("*" * 83)
display_html( temp/ np.sqrt((temp**2).sum()) )
print("\n")

print("*" * 50)
print("     If you insist on using scikit learn then")
print("*" * 50)
print("\n")

l2 = Normalizer(norm="l2")
print(l2.fit_transform(temp.T).T)
```

## Min Max regularization
Each random variables can be negative or positive
$$\mathbb{Y} = \{ y_n \in \mathbb{R} \}$$

$$\mathbb{X} = \dfrac{y_n - min( \mathbb{Y} )} {max( \mathbb{Y} ) - min( \mathbb{Y} )}$$

After normalization
$$\mathbb{X} = \{ x_n \in \mathbb{R} : 0 \leq y_n \leq 1 \}$$

```python
print("*" * 83)
print(" Scikit learn does it WRONG where it applies it row wise which is definitely wrong")
print("*" * 83)
print("\n")
l_max = Normalizer(norm="max")
print(l_max.fit_transform(temp))
print("\n")

print("*" * 83)
print("   THE WRIGHT WAY. Also luckily pandas does column wise computation by default")
print("*" * 83)
display_html( (temp - temp.min()) / (temp.max() - temp.min()) )
print("\n")

print("*" * 50)
print("     If you insist on using scikit learn then")
print("*" * 50)
print("\n")

l_max = Normalizer(norm="max")
print(l_max.fit_transform(temp.T).T)
```

## Scaling around mean and standard deviation
Each random variables can be negative or positive
$$\mathbb{Y} = \{ y_n \in \mathbb{R} \}$$

After scaling
$$\color{blue}{Y \sim Normal \; (\mu=0,\sigma^2=1)}$$

```python
print("*" * 100)
print("ddof = 0 to treat it as Population otherwise 1(default) to treat it as sample meaning divide by N-1")
print("*" * 100)
display_html((temp - temp.mean()) / temp.std(ddof=0))
print("\n")

print("*" * 90)
print("By default scikit-learn treats random variables as population")
print("*" * 90)
StandardScaler().fit_transform(temp)
```
