## Let's see what KMeans think how they should be clustered

```r
iris.kmeans <- kmeans(iris[,c("Sepal.Length","Sepal.Width")],3)
```

```r
plot(iris$Sepal.Length, iris$Sepal.Width, col=iris.kmeans$cluster, pch=19)
legend("topright", legend=unique(iris.kmeans$cluster), col=c("black", "red","green"), lty=1:2, cex=0.8)
points(iris.kmeans$centers, pch="X", col="blue", cex=2)
```
