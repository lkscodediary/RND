## Test for Autocorrelation
| Test | Null Hypothesis | Interpretation |
|------|----------------|----------------|
| Ljung-Box | No autocorrelation up to lag k | p < 0.05: autocorrelation exists |
| Box-Pierce | No autocorrelation up to lag k | p < 0.05: autocorrelation exists |

```r
library(tseries)
library(ggplot2)
```

## Helper Function

```r
autocorrelation_test <- function(ts_vec, lags = 20) {
    results <- list()

    # Ljung-Box
    lb_res  <- Box.test(ts_vec, lag=lags, type="Ljung-Box")
    # Box-Pierce
    bp_res  <- Box.test(ts_vec, lag=lags, type="Box-Pierce")

    stationary <- lb_res$p.value > 0.05 && bp_res$p.value > 0.05
    msg <- if (!stationary)
        "Strong evidence of serial autocorrelation (non-stationary)"
    else
        "No significant autocorrelation detected (stationary)"
    cat("Ljung-Box:", msg, "\n")

    data.frame(
        Test       = c("Ljung-Box","Box-Pierce"),
        Statistic  = c(round(lb_res$statistic,4), round(bp_res$statistic,4)),
        P.Value    = c(round(lb_res$p.value,4),   round(bp_res$p.value,4)),
        Lags       = lags,
        Stationary = if (stationary) "Yes" else "No",
        Details    = msg
    )
}
```

## Time Series in levels

```r
usd_cad <- read.csv("../../../data/USD_CAD.csv")
rownames(usd_cad) <- usd_cad$time
```

## USD_CAD

```r
n_plot <- 500
par(mfrow=c(2,1), mar=c(2,4,2,1))
acf(tail(usd_cad$m_c, n_plot),  lag.max=50, main="ACF - USD/CAD Level")
pacf(tail(usd_cad$m_c, n_plot), lag.max=50, main="PACF - USD/CAD Level")
par(mfrow=c(1,1))

plot(usd_cad$m_c, type="l", main="USD/CAD Level", ylab="Rate")
grid()
```

```r
autocorrelation_test(usd_cad$m_c, lags=20)
```

## Conclusion
- USD/CAD in levels shows clear autocorrelation

## USD_CAD - First order log difference

```r
ts_diff <- diff(log(usd_cad$m_c))
n_plot  <- 500
par(mfrow=c(2,1), mar=c(2,4,2,1))
acf(tail(ts_diff, n_plot),  lag.max=50, main="ACF - Log Difference")
pacf(tail(ts_diff, n_plot), lag.max=50, main="PACF - Log Difference")
par(mfrow=c(1,1))

plot(ts_diff, type="l", main="USD/CAD Log Difference", ylab="Log Return")
grid()
```

```r
autocorrelation_test(ts_diff, lags=20)
```

## Conclusion
- USD/CAD in log difference shows **no** clear autocorrelation
