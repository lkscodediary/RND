## Regularization
Scaling features to similar magnitudes for better optimiser convergence.

Methods:
- **L1**: Divide by sum of absolute values
- **L2**: Divide by square root of sum of squares
- **Min-Max**: Scale to [0, 1]
- **Standard Scaler**: Subtract mean, divide by std → standard normal

```r
# No extra packages needed — base R and scale() handle everything
library(knitr)  # for nice table display
```

## Sample Data

```r
temp <- data.frame(
    A = c(4, 1, 2),
    B = c(1, 3, 2),
    C = c(2, 5, 5),
    D = c(2, 3, 3)
)
print(temp)
```

## L1 Regularization
Divide each column by the sum of absolute values of that column.
$$\omega = \sum_{n=0}^{N} |y_n|$$

```r
cat(strrep("*",20), "\n")
cat("   THE RIGHT WAY (column-wise)\n")
cat(strrep("*",20), "\n")
l1_normalized <- sweep(temp, 2, colSums(abs(temp)), FUN="/")
print(l1_normalized)
```

## L2 Regularization
Divide by the square root of the sum of squares of each column.
$$\omega = \sqrt{\sum_{n=0}^{N} y_n^2}$$

```r
l2_norms      <- sqrt(colSums(temp^2))
l2_normalized <- sweep(temp, 2, l2_norms, FUN="/")
cat(strrep("*",83), "\n")
cat("   THE RIGHT WAY (column-wise)\n")
cat(strrep("*",83), "\n")
print(l2_normalized)
```

## Min-Max Regularization
$$x_n = \frac{y_n - \min(Y)}{\max(Y) - \min(Y)}$$

```r
min_vals <- apply(temp, 2, min)
max_vals <- apply(temp, 2, max)
minmax   <- sweep(sweep(temp, 2, min_vals, "-"), 2, max_vals - min_vals, "/")
cat(strrep("*",83), "\n")
cat("   THE RIGHT WAY (column-wise)\n")
cat(strrep("*",83), "\n")
print(minmax)
```

## Standard Scaling (around mean and standard deviation)
After scaling: $Y \sim Normal(\mu=0, \sigma^2=1)$

```r
cat(strrep("*",100), "\n")
cat("ddof=0 (population std) vs ddof=1 (sample std, R default)\n")
cat(strrep("*",100), "\n")

# Population std (ddof=0): divide by N
pop_std    <- apply(temp, 2, function(x) sd(x) * sqrt((length(x)-1)/length(x)))
scaled_pop <- sweep(sweep(temp, 2, colMeans(temp), "-"), 2, pop_std, "/")
cat("\nPopulation std scaling (equivalent to sklearn StandardScaler):\n")
print(scaled_pop)

cat("\nSample std scaling (R default, ddof=1):\n")
print(scale(temp))
```
